```python
# making dataset that is usable, sum over blocknum, merge with category
data = salesTrain.groupby(['date_block_num', 'shop_id', 'item_id', 'item_price'], as_index=False)['item_cnt_day'].sum() # sum item sales
data = data.set_index('item_id').join(items[['item_category_id', 'item_id']].set_index('item_id'))
data = data.reset_index()

corrMatrix = data.corr()
sn.heatmap(corrMatrix, annot=True, cmap='Reds')
plt.title("Correlation matrix of salesTrain")
plt.show()
```

```python
# X=date_block_num, shop_id, item_id Y=item_cnt_day
# could also include price
from sklearn.model_selection import train_test_split

X = data[['date_block_num', 'shop_id', 'item_id']].values
X2 = data[['date_block_num', 'shop_id', 'item_id', 'item_category_id']].values
X3 = data[['date_block_num', 'shop_id', 'item_id', 'item_category_id', 'item_price']].values
X4 = data[['date_block_num', 'shop_id', 'item_id', 'item_price']].values
Y = data['item_cnt_day'].values
Y = Y.reshape(-1,1)
print(X, X2, X3)
print(Y)
print(len(Y))

X_train, X_test, y_train, y_test = train_test_split(X4, Y, test_size=0.3, random_state=0)
```

### linear regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(normalize=True)
model.fit(X_train, y_train)
print(model.score(X_test,y_test))
```

```python
# param choose
from sklearn.model_selection import GridSearchCV
def TestParams(model, params, nTest):
    global X_train, y_train
    global X_test, y_test
    
    knnGr = GridSearchCV(model, params, cv=nTest, verbose=True, n_jobs=8)
    knnGr.fit(X_train, y_train)
    print(knnGr.score(X_test, y_test))
    return pd.DataFrame(knnGr.cv_results_)
```

```python
# SVR, support vector regression
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

a = np.array([])
for i in range(10):
    print(i)
    regr = make_pipeline(StandardScaler(), SGDRegressor(max_iter=1000, tol=1e-3))
    regr.fit(X_train, y_train)
    a = np.append(a, regr.score(X_test, y_test))
print(a.mean())
```

```python
#Decision tree
from sklearn.tree import DecisionTreeRegressor 

mytree = DecisionTreeRegressor(max_depth=17)
mytree.fit(X_train, y_train)
print(mytree.score(X_test, y_test))
```

```python
#extra tree
from sklearn.ensemble import ExtraTreesRegressor
for d in np.arange(25, 30, 1):
    exTree = ExtraTreesRegressor(n_estimators=40, max_depth=d, n_jobs=4)
    exTree.fit(X_train, y_train)
    print(d, exTree.score(X_test, y_test)) # 0.5750281515370448
```

```python
#Random forest
from sklearn.ensemble import RandomForestRegressor
randForr = RandomForestRegressor(max_depth=32, random_state=0, n_jobs=4)
randForr.fit(X_train,y_train)
print(randForr.score(X_test, y_test))
```

```python
#Polynomial regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

print("poly")
degree = 9
polyDeg=make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyDeg.fit(X_train, y_train)
print(polyDeg.score(X_test, y_test))
```

```python
# KNN regressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV

knn = KNeighborsRegressor(n_jobs=3)
params = {'n_neighbors': np.arange(1,40)}
knnGr = GridSearchCV(knn, params, cv=2, verbose=True, n_jobs=8)
knnGr.fit(X_train, y_train)
print(knnGr.score(X_test, y_test))
```

```python
res = pd.DataFrame(knnGr.cv_results_)
print(res)
plt.plot(res['mean_test_score'].values)
plt.xlabel("Neighbors")
plt.ylabel("R^2 score")
plt.title("KNN regression")
plt.show()
print(knnGr.best_params_)
```

```python
# AI
import tensorflow.keras as keras
from sklearn import datasets, model_selection
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Conv2D
from sklearn.metrics import r2_score
import tensorflow as tf
from keras.callbacks import EarlyStopping

model = Sequential()
model.add(Dense(512, activation='tanh', input_shape=(X_train[0].shape[0],)))
model.add(Dense(512, activation='relu'))
model.add(Dense(512, activation='sigmoid'))
model.add(Dense(512, activation='sigmoid'))
model.add(Dense(1, activation='linear'))

overfitCallback = EarlyStopping(monitor='loss', min_delta=0, patience=1)

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])

model.fit(X_train, y_train, callbacks=[overfitCallback], verbose=True, epochs=10)

```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, model.predict(X_test)))
```
